```python
import pandas as pd
import numpy as np
import requests, time
from bs4 import BeautifulSoup
import sqlalchemy

pd.set_option('display.max_columns', None)  
```

```python
url = 'http://pfref.com/pi/share/D42i8'
```

```python
print(url)
```

```python
dl= pd.read_html(url)
print(len(dl))
```

```python
dl
```

```python
dt = dl[0]
dt
```

```python
dt = dt.rename(columns={'Unnamed: 0_level_0':'',
                        'Unnamed: 1_level_0': '', 
                        'Unnamed: 2_level_0': '',
                        'Unnamed: 3_level_0': '',
                        'Unnamed: 4_level_0': ''  })
```

```python
dt
```

```python
dt.loc[:,'All-Pro'] = dt['','Player'].str.contains('+', regex=False)
dt.loc[:,'Selected to Pro Bowl'] = dt['','Player'].str.contains('*', regex=False)
```

```python
dt
```

```python
dt = dt.dropna(how='all')
```

```python
dt
```

```python
dt['','Player'] = dt['', 'Player'].str.replace('*','')
dt['','Player'] = dt['', 'Player'].str.replace('+','')
```

```python
dt
```

```python
dt.columns = [(c.lower().strip() + ' '+ d.lower().strip()).strip() for c,d in dt.columns]
```

```python
db_server='pymysql'
user='root'
db_port = '3306'
password = ''
ip = 'localhost'
db_name = 'nflcopy'
engine = sqlalchemy.create_engine(f'mysql+{db_server}://{user}:{password}@{ip}:{db_port}/{db_name}?charset=utf8')
```

```python
dt.to_sql('kicker', engine, index=False, if_exists='replace')
```
