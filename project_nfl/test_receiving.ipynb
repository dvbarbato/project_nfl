```python
def get_data(url):
    response = requests.get(url)
    html = response.content
    
    return html
```

```python
get_data(url)
```

```python
def get_headers(soup):
    headers = soup.findAll('tr')[0]
    headers = [i.getText() for i in headers.findAll('th')]
    
    return headers
```

```python
get_headers(soup)
```

```python
def setup(url):
    pure_data = get_data(url)
    soup = BeautifulSoup(pure_data)
    col_headers = get_headers(soup)
    rows = soup.find_all('tr')[1:]
    
    return rows
```

```python
setup(url)
```

```python
def populate_rec(rows):
    rec_stats = []
    for i in range(len(rows)):
        rec_stats.append( [col.getText() for col in rows[i].find_all('td')])
        
    return rec_stats
```

```python
populate_rec(rows)
```

```python
def init_dataframe(rec_stats, column_headers):    
    data = pd.DataFrame(rec_stats, columns=column_headers[1:])
    
    return data
```

```python
init_dataframe(rec_stats, column_headers)
```

```python
def sanitizer_rec(data):
    data = data.dropna()
    data.columns = [column.lower().strip() for column in data.columns]
    data['player'] = data['player'].str.replace('*', '')
    data['player'] = data['player'].str.replace('+', '')
    
    return data
```

```python
sanitizer_rec(data)
```

```python
def robos(url):
    page_html = get_data(url)
    soup = BeautifulSoup(page_html)
    name_cols = get_headers(soup)
    contents_row = soup.find_all('tr')[1:]
    traits_rec = populate_rec(contents_row)
    
    table = init_dataframe(traits_rec, name_cols)
    return sanitizer_rec(table)
```

```python
robos(url)
```

```python
df2 = robos(url)
```
